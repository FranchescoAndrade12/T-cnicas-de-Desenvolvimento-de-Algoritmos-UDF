```python
A = [[4, 5],
     [2, 3]]

B = [[3, 6],
     [3, 9]]

C = [[0, 0],
     [0, 0]]

print("Matriz:")
print(A)
print(B)

for i in range(2):
  for j in range(2):
    C [i] [j] = A [i] [j] + B [i] [j]

print("A soma é:", C)

D = [[0, 0],
     [0, 0]]

for i in range(2):
  for j in range(2):
    D [i] [j] = A [i] [j] - B [i] [j]

print("A subtração é:", D)

E = [[0, 0],
     [0, 0]]

for i in range(2):
  for j in range(2):
    for k in range(2):
      E [i] [j] += A [i] [k] * B [k] [j]

print("A multiplicação é: ", E)
```
